### Using Langchain Model - openai

#### Load env variables

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model= "gpt-4.1-2025-04-14", temperature=0)
```

#### using .invoke

```python
result =llm.invoke("when was the last bank of 2025 in Ireland?")
print(result.content)
```
